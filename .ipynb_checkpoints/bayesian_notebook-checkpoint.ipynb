```python
train = titanic[:700]
test = titanic[700:]
```

```python
with pm.Model() as log_model:

    a = pm.Normal(name='intercept',mu = 0.0, tau = 1/float(10)**2)
    b_class = pm.Normal(name='b_class',mu = 0.0, tau = 1/float(10)**2)
    b_sex = pm.Normal(name='b_sex', mu = 0.0, tau = 1/float(10)**2)
    b_sibling = pm.Normal(name='b_sibling', mu = 0.0, tau = 1/float(10)**2)
    b_child = pm.Normal(name='b_child', mu = 0.0, tau = 1/float(10)**2)
    b_fare = pm.Normal(name='b_fare', mu = 0.0, tau = 1/float(10)**2)
    b_age = pm.Normal(name='b_age', mu = 0.0, tau = 1/float(10)**2)
    
    yhat = a + (b_class * train['Pclass']) + (b_sex * train['Sex_numeric']) + b_sibling * train['Siblings/Spouses Aboard'] + b_child * train['Parents/Children Aboard'] + b_fare * train['Fare'] + b_age * train['Age']
    #yhat = a + b_age * train['Age'] + b_fare * train['Fare'] +b_sex * train['Sex_numeric'] + b_fare * train['log_fare']
    theta = pm.math.sigmoid(yhat)
    y = pm.Bernoulli('y', p =theta, observed=train['Survived'].values)
    
```

```python
with log_model:
    trc_fish = pm.sample(10000, chains=3)
```

```python
trc_fish = trc_fish[1000:]
```

```python
pm.traceplot(trc_fish, var_names=['b_age', 'b_class', 'b_sex', 'b_sibling', 'b_child', 'b_fare'])
```

```python
pm.autocorrplot(trc_fish)
```

```python
yhat = trc_fish['intercept'].mean() + trc_fish['b_age'].mean() * test['Age'] + trc_fish['b_class'].mean() * test['Pclass'] + trc_fish['b_sex'].mean() * test['Sex_numeric'] + trc_fish['b_sibling'].mean() * test['Siblings/Spouses Aboard'] + trc_fish['b_child'].mean() * test['Parents/Children Aboard'] + trc_fish['b_fare'].mean() * test['Fare']
    
test_pred = 1 / (1 + np.exp(-1 *(yhat)))
```

```python
yhat = trc_fish['intercept'].mean() + trc_fish['b_age'].mean() * train['Age'] + trc_fish['b_class'].mean() * train['Pclass'] + trc_fish['b_sex'].mean() * train['Sex_numeric'] + trc_fish['b_sibling'].mean() * train['Siblings/Spouses Aboard'] + trc_fish['b_child'].mean() * train['Parents/Children Aboard'] + trc_fish['b_fare'].mean() * train['Fare']
    
train_pred = 1 / (1 + np.exp(-1 *(yhat)))
```

```python
train_Accuracy =((train_pred>0.5).astype(np.int) == train['Survived']).mean()
model_accuracy = ((test_pred>0.5).astype(np.int) == test['Survived']).mean()
print(train_Accuracy, model_accuracy)
```
